## Using pandas datetime properties

```python
air_quality["datetime"] = pd.to_datetime(air_quality["datetime"])
```

```python
air_quality["datetime"]
```

 By applying the to_datetime function, pandas interprets the strings and convert these to datetime (i.e. datetime64[ns, UTC]) objects.

this is similar to datetime.datetime from standard library as pandas.Timestamp

NOTE:
As many data sets do contain datetime information in one of the columns, pandas input function like pandas.read_csv() and pandas.read_json() can do the transformation to dates when reading the data using the parse_dates parameter with a list of the columns to read as Timestamp:

pd.read_csv("../data/air_quality_no2_long.csv", parse_dates=["datetime"])

```python
air_quality["datetime"].min(), air_quality["datetime"].max()
```

```python
air_quality["datetime"].max() - air_quality["datetime"].min()
```

To add a new column to the DataFrame containing only the month of the measurement

```python
air_quality['month'] = air_quality['datetime'].dt.month
```

```python
air_quality.head()
```

By using Timestamp objects for dates, a lot of time-related properties are provided by pandas. For example the month, but also year, quarter,… 

All of these properties are accessible by the dt accessor.

##### average NO2
 concentration for each day of the week for each of the measurement locations?

```python
air_quality.groupby([air_quality['datetime'].dt.weekday, "location"])["value"].mean()
```

To group on weekdays, we use the datetime property weekday (with Monday=0 and Sunday=6) of pandas Timestamp

```python
fig, axis = plt.subplots(figsize=(12,4))

air_quality.groupby(air_quality['datetime'].dt.hour)["value"].mean().plot(
    kind='bar', rot=0, ax=axs
)
```

```python
plt.xlabel("Hour of the day");  #custom x label using Matplotlib

plt.ylabel("$NO_2 (µg/m^3)$")
```

## Resample a time series to another frequency

```python
monthly_max = no_2.resample("M").max()
```

```python
monthly_max
```

The resample() method is similar to a groupby operation:

- it provides a time-based grouping, by using a string (e.g. M, 5H,…) that defines the target frequency

- it requires an aggregation function such as mean, max,…

```python
monthly_max.index.freq
```

a plot of the daily mean NO2
 value in each of the stations.
