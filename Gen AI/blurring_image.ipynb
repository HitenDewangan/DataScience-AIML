```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('img.jpg')

# Apply Gaussian Blur
gaussian_blur = cv2.GaussianBlur(img, (25, 25), 0)

# Apply Box Blur
box_blur = cv2.blur(img, (25, 25))

# Apply Median Blur
median_blur = cv2.medianBlur(img, 25)

# Convert images to RGB format for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gaussian_blur_rgb = cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB)
box_blur_rgb = cv2.cvtColor(box_blur, cv2.COLOR_BGR2RGB)
median_blur_rgb = cv2.cvtColor(median_blur, cv2.COLOR_BGR2RGB)

# Create a subplot
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Display the original and blurred images
axs[0, 0].imshow(img_rgb)
axs[0, 0].set_title('Original')
axs[0, 1].imshow(gaussian_blur_rgb)
axs[0, 1].set_title('Gaussian Blur')
axs[1, 0].imshow(box_blur_rgb)
axs[1, 0].set_title('Box Blur')
axs[1, 1].imshow(median_blur_rgb)
axs[1, 1].set_title('Median Blur')

# Remove axis ticks
for ax in axs.flat:
    ax.set_xticks([])
    ax.set_yticks([])

# Show the plot
plt.show()

# Save the blurred images
# cv2.imwrite('gaussian_blur.jpg', gaussian_blur)
# cv2.imwrite('box_blur.jpg', box_blur)
# cv2.imwrite('median_blur.jpg', median_blur)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('img.jpg')

# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the original and grayscale images
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original')
axs[1].imshow(gray_img, cmap='gray')
axs[1].set_title('Grayscale')

# Remove axis ticks
for ax in axs.flat:
    ax.set_xticks([])
    ax.set_yticks([])

# Show the plot
plt.show()

# # Save the grayscale image
# cv2.imwrite('grayscale_image.jpg', gray_img)
```
