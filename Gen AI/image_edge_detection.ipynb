```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image 'house.jpeg'
def read_image(path_to_image):
    image = cv2.imread(path_to_image)
    return image

# Show the image
def show_image(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()

# Generate edges of the image using Canny Edge Detector
def generate_canny_edge(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
    canny_edges = cv2.Canny(blurred_image, 50, 150)
    return canny_edges

# Main function
def main():
    path_to_image = 'house.jpeg'
    image = read_image(path_to_image)
    show_image(image)
    canny_edges = generate_canny_edge(image)
    plt.imshow(canny_edges, cmap='gray')
    plt.show()

if __name__ == "__main__":
    main()
```
